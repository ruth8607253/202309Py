## git指令
`git clone "網址"`：用clone的網址下載GitHub的repo。

範例：`git clone https://github.com/ruth8607253/Flask.git`

![git1](./image/git1.jpg)

`git status`

`git config --list`

`git remote -v`：看這裡有沒有連到遠端

`git add 檔案名稱`：將「工作區(working area)」已追蹤的檔案，放到「暫存區(stage)」

**vscode介面**-在「原始檔控制」裡，把原本放在「變更」的檔案，放到「暫存的變更」

`git commit -m "我做了什麼事情"`：描述檔案修改內容

`git commit -a`：請電腦偵測更改的檔案(除了新增的檔案)，將它們新增到索引，並提交。

`git push`：將「暫存區(stage)」的檔案上傳(commit)到儲存庫(repositories，簡稱：repo)

**vscode介面**-按「提交」

`git push --force`：(以電腦上的資料為主)，把雲端的資料覆蓋掉。

發生情境：在雲端上有一個資料A，但我忘了，在電腦上也建了一個資料A。當我要上船電腦上的資料A時就會有問題(錯亂)。

`git init`：新增一個.git在資料夾裡面。移除：rm -

## 用git在vscode恢復上一棟：Sourcetree
### 創建資料與上傳
#### 創建資料夾 & 檔案
1. 創建資料夾：

    在自己想開資料夾的地方(e.g.桌面)創一個資料夾 -> 用vscode開啟資料夾a

2. 設定Sourcetree

    點上面按鈕「Add」 -> 選取資料夾 -> Add

    ![sourcetree5](./image/sourcetree5.jpg)

3. 創建檔案：`touch 檔案名稱.類型` 如：touch a1.py

4. 在檔案裡打些東西：`vim a1.py` -> print("Hello, AAA!") -> `Esc` + `:` + `wq`
#### 上傳資料
1. 將檔案放到暫存區：`git add .`

2. 打新增/修改備註：`git commit -m "新增/修改備註"`  如：git commit -m "創a1"

3. 上傳：`git push --set-upstream origin master`

    --set-upstream = -u ：以後只要打「git push」(不用再打git push origin master)就可以直接上傳

這樣在Sourcetree就能看到修改內容：

打開Sourcetree -> 點左邊選單的master(空心圓點那個) -> 顯示跟GitHub的Commit紀錄很像的東西。

![sourcetree6](./image/sourcetree6.jpg)
#### 將Commit移回暫存(Unstage File)
`git reset --soft`
#### 將Commit移回變更(Stage File)
1. 在終端機打`git reset --mixed`
2. 打開Sourcetree -> 點左邊選單的master(空心圓點那個) -> 對修改
#### 將Commit完全取消
`git reset --hard`
#### 範例
`git reset --hard^`：回上1棟

`git reset --hard^^`：回上2棟

以此類推

## lenux指令
### 目錄與位置
`./`：當前目錄

`../`：上一層目錄

`~`：回到個人工作目錄

`cd`：change directery 更換目錄位置

絕對路徑寫法：cd 把要打開的資料夾位置完整打出來

範例：cd /c/Users/user/Documents/GitHub/202309Py

相對路徑寫法： cd ./或../過去想要打開的資料夾

範例：原本在202309Py/a資料夾裡，要換去202309Py/b資料夾 -> cd ../ b

`pwd`：查看檔案來源

`ls`：list列出當前目錄內容

`ls-al`：list-all查看所有目錄(可以看到隱藏檔)
### 創建、移動與複製檔案&資料夾
`touch`建立檔案

範例：touch a.txt

`mv`：move 移動 mv 要移動的檔案 要移動到這邊

範例：我要移動202309Py/a資料夾裡**所有的東西**，到202309Py/b資料夾裡 -> mv ./a/a1.txt ./b/

*不能直接把repoA移到repoB。

因為每個repo裡都有一個.git(隱藏檔)，.git不同存在於同一個資料夾裡。

`rm`：remove 移除

`cp`：copy 複製 cp 要複製的檔案 要複製到這邊

範例：我要複製202309Py/a資料夾裡的a1，到202309Py/b資料夾裡 -> cp ./a/a1.txt ./b/

`mkdir`：make 

`nano`：Unix的文字編輯器，可以在終端機編輯檔案。

離開：`Ctrl` + `x`，會問你要不要存檔(y/n)

範例：用終端機在aaa.py裡打print("Hello, world!") -> `Ctrl` + `x`，要存檔按`y`

nano aaa.py -> 打print("Hello, world!") -> `Ctrl` + `x`，要存檔按`y`

`vim`：在終端機新增+編輯檔案。

存檔+出來：`Esc` + `:` + `wq`

範例：在終端機打vim a1.py -> 在裡打print("Hello, world!") -> `Esc` + `:` + `wq`

vim a1.py -> 打print("Hello, world!") -> `Esc` + `:` + `wq`
