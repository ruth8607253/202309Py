```python
import random
import statistics

class Player:
    def __init__(self, name, d1=0, d2=0, d3=0, d4=0):
        self.__dice1 = d1
        self.__dice2 = d2
        self.__dice3 = d3
        self.__dice4 = d4
        self.name = name

    def __play(self) -> int:
        Dice = [self.__dice1, self.__dice2, self.__dice3, self.__dice4]

        if len(set(Dice)) == 1:
            if sum(Dice) == 4:
                score = 13
            elif sum(Dice) == 8:
                score = 14
            elif sum(Dice) == 12:
                score = 15
            elif sum(Dice) == 16:
                score = 16
            elif sum(Dice) == 24:
                score = 17
            else:
                score = 18
        elif len(set(Dice)) == 4 or Dice.count(self.__dice1) == 3 or Dice.count(self.__dice2) == 3 or Dice.count(
                self.__dice3) == 3 or Dice.count(self.__dice4) == 3:
            score = 0
        else:
            score = sum(Dice) - min(statistics.multimode(Dice)) * 2

        return score

    @property
    def value(self) -> int:
        return self.__play()

    def __repr__(self) -> str:
        descript = ""
        descript += f"{self.name}\n"
        descript += f"骰子1={self.__dice1}\n"
        descript += f"骰子2={self.__dice2}\n"
        descript += f"骰子3={self.__dice3}\n"
        descript += f"骰子4={self.__dice4}\n"
        descript += f"得分={self.value}分"
        return descript

def main():
    p1 = Player("robert")
    p2 = Player("john")

    while True:
        p1.__dict__['__dice1'] = random.randint(1, 6)
        p1.__dict__['__dice2'] = random.randint(1, 6)
        p1.__dict__['__dice3'] = random.randint(1, 6)
        p1.__dict__['__dice4'] = random.randint(1, 6)

        p2.__dict__['__dice1'] = random.randint(1, 6)
        p2.__dict__['__dice2'] = random.randint(1, 6)
        p2.__dict__['__dice3'] = random.randint(1, 6)
        p2.__dict__['__dice4'] = random.randint(1, 6)

        print(f"{p1.name} 的骰子结果: {p1.__dict__['__dice1']}, {p1.__dict__['__dice2']}, {p1.__dict__['__dice3']}, {p1.__dict__['__dice4']}")
        print(p1.value)
        print(p1)

        print(f"{p2.name} 的骰子结果: {p2.__dict__['__dice1']}, {p2.__dict__['__dice2']}, {p2.__dict__['__dice3']}, {p2.__dict__['__dice4']}")
        print(p2.value)
        print(p2)

        user_input = input("按 Enter 键继续游戏，按 'q' 键退出游戏: ")
        if user_input == 'q':
            break

    print("游戏结束。")

if __name__ == "__main__":
    main()
```
